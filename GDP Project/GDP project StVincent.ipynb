```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
vincent_df=pd.read_csv("St Vincent GDP by sector.csv")
```

```python
vincent_df
```

```python
vincent_df = vincent_df.rename(columns={'2021 P': '2021'})
```

```python
vincent_df.head()
```

```python
vincent_temp = vincent_df
vincent_temp=vincent_temp.T
##removing the indexes
vincent_temp.reset_index(inplace=True)
print(vincent_temp.columns)
```

```python
vincent_temp
```

```python
years=vincent_temp['index']
years=pd.DataFrame(years)
years.columns=["Years"]
years=years[1:]
years.T
```

```python
vincent_temp=vincent_temp.drop(['index'],axis=1)
vincent_temp.T
vincent_temp.columns=vincent_temp.iloc[0]
```

```python
vincent_temp
vincent_temp["Year"]=years
vincent_temp.T
vincent_temp.head()
```

```python
clean_vincent=vincent_temp[1:]
clean_vincent = clean_vincent[['Year','Agriculture, forestry and fishing' , 'Mining and quarrying', 'Manufacturing', 'Electricity, gas, steam and air conditioning supply', 'Water supply, sewerage, waste management and remediation activities', 'Construction', 'Wholesale and retail trade; repair of motor vehicles and motorcycles', 'Transportation and storage', 'Accommodation and food service activities', 'Information and communication', 'Financial and insurance activities', 'Real estate activities', 'Professional, scientific and technical activities', 'Administrative and support service activities', 'Public administration and defence; compulsory social security', 'Education', 'Human health and social work activities', 'Arts, entertainment and recreation', 'Other service activities', 'Gross Value Added at Basic Prices', 'GROWTH RATE']]
clean_vincent.head()
```

```python
clean_vincent = clean_vincent.iloc[:, :-1]
```

```python
for index, value in clean_vincent['Gross Value Added at Basic Prices'].items():
    clean_vincent.loc[index, 'Gross Value Added at Basic Prices'] = value.replace(',', '')
```

```python
clean_vincent.head()
```

```python
print( clean_vincent.isnull().any())
```

```python
numeric_cols = clean_vincent.columns.drop('Year')
clean_vincent[numeric_cols] = clean_vincent[numeric_cols].astype(float)
```

```python
# Set the figure size
plt.figure(figsize=(10, 15))

# Define a list of colors for the lines
colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow', 'cyan', 'magenta', 'lime', 'pink', 'brown', 'teal', 'navy', 'indigo', 'darkgreen', 'olive', 'salmon', 'sienna', 'lightblue']

# Plotting the graph with different colors for each line
for col, color in zip(clean_vincent.columns[1:], colors):
    plt.plot(clean_vincent['Year'], clean_vincent[col] , marker='o', label=col, color=color)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Value in EC$ Million')
plt.title('Sector Contribution to St.Vincent and Grenadines GDP')

# Place the legend outside the plot area
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='small')

plt.grid(True)  # Add grid lines
plt.xticks(clean_vincent['Year'])  # Set x-axis ticks to match years
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
# Plotting the scatter plot
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

# Plotting the Manufacturing sector on the first subplot
ax1.plot(clean_vincent['Year'], clean_vincent['Wholesale and retail trade; repair of motor vehicles and motorcycles'], marker='o', color='green', label='Wholesale and retail trade; repair of motor vehicles and motorcycles')
ax1.set_title('Wholesale and retail trade; repair of motor vehicles and motorcycles')
ax1.set_xlabel('Year')
ax1.set_ylabel('Value in EC$ Million')
ax1.legend()
ax1.grid(True)

# Plotting the Trade and repairs sector on the second subplot
ax2.plot(clean_vincent['Year'], clean_vincent['Real estate activities'], marker='o', color='#FF1493' , label='Real estate activities')
ax2.set_title('Real Estate, Renting and Business Activities Sector')
ax2.set_xlabel('Year')
ax2.set_ylabel('Value in EC$ Million')
ax2.legend()
ax2.grid(True)

# Plotting the Mining and quarrying sector on the third subplot
ax3.plot(clean_vincent['Year'], clean_vincent['Public administration and defence; compulsory social security'], marker='o', color='orange', label='Public administration and defence; compulsory social security')
ax3.set_title('Public administration and defence; compulsory social security')
ax3.set_xlabel('Year')
ax3.set_ylabel('Value in EC$ Million')
ax3.legend()
ax3.grid(True)


# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
