All library imports that I've used to scrape and clean the data

```python
#816030787----------
#Faith Shim---------
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import pandas as pd
```

 After inspecting the URL of the site I noted the only change for each page was at the end with a month and year string. Using this to my advantage I created a loop that would loop through the days of each year and scraped the data from each page. Being careful with NULL values or pages with no data.

>  The current method works as it's only 13 years of data however if it were more it may be more viable to have two arrays and do a nested for loop to ensure accuracy.



```python
baseURL = "https://www.nlcbplaywhelotto.com/nlcb-lotto-plus-results/?monthyear="

startDate = datetime.strptime("Jan-10", "%b-%y")
endDate = datetime.strptime("Jan-24", "%b-%y")

currentDate = startDate;
addDays = timedelta(days=31)

lotto_results = []
while currentDate < endDate:
  dateText= currentDate.strftime("%b-%y")
  URL = f"{baseURL}{dateText}"
  page = requests.get(URL)
  soup = BeautifulSoup(page.content, "html.parser")
  results = soup.find(id="monthResults")
  if results is not None:
    lotto_results.append(results)

  currentDate += addDays;

```

The following two blocks of code pull in the raw date values and the raw information values below the dates into two separate arrays

```python
raw_dates = []
for result in lotto_results:
  raw_dates.append(result.find_all('tr', class_="lotto-date-tr" ))
```

```python
raw_values = []

for result in lotto_results:
  raw_values.append(result.find_all('tr', class_= "lotto-tr"))
```

The following two blocks cleans up the raw values by stripping away unnecessary text to leave behind the data we will use

```python
clean_values=[]
for value in raw_values:

  for data in value:
    data_val = data.find_all('td')
    clean_data =[val.text.strip() for val in data_val]

    if clean_data:

      clean_values.append(clean_data)



print(clean_values)
```

```python
clean_dates=[]
for value in raw_dates:

  for data in value:
    data_val = data.find_all('td')
    clean_data =[val.text.strip() for val in data_val]

    if clean_data:
       clean_dates.append(clean_data)



print(clean_dates)
```

This block merges the clean_dates and clean_values to form clean_data

```python
clean_data = []

for x in range(0, len(clean_dates),1):
  clean_data.append(clean_dates[x]+ clean_values[x])

print(clean_data)
```

Create list that holds all the header names and then add date to the front of the list

```python
for result in lotto_results:
  col_names = result.find_all('th')

col_names = [name.text for name in col_names]
col_names.insert(0,'Date')

print(col_names)
```

Using pandas we create a lotto dataframe and fill it with our clean_data.
Then we change the column names to the ones we created previously

```python
lotto = pd.DataFrame(clean_data)
lotto.columns = col_names
```

Reviewing the website and clean_data we note that there are some X values in the data

```python
lotto.eq('X').sum()
```

There are as many X values in Jackpot as in Wins, showing a correlation to the lack of a jackpot meaning the lack of a winner value. These rows would not give alot of information towards most research based on the Jackpot values and thus I removed it at my discretion

```python
lotto = lotto.drop(lotto[lotto.Wins == 'X'].index)
lotto.eq('X').sum()
```

Reviewing the website and clean_data we note that there are some NA values in the data

```python
lotto.eq('NA').sum()
```

We can make the assumption that multipliers were not added to the game yet and thus the "multiplier" would be 1 as it would just be the jackpot value itself. Then ensure no more NA values exist

```python
lotto.replace('NA','1', inplace=True)
lotto.eq('NA').sum()
```

Convert Numbers column in lotto into a string of num values

```python
num_lotto = []
for num in lotto['Numbers']:
  num_lotto = lotto['Numbers'].str.split('-')

lotto['Numbers'] = num_lotto
print(lotto['Numbers'])
```

Convert date from 02-Jan-10 format to 02-01-2010 format.

---


Note if this runs once then it shouldnt be run again as the date was updated and would no longer be in the format %d-%b-%y and the function won't work

```python
lotto['Date'] = pd.to_datetime(lotto['Date'],format ='%d-%b-%y' ).dt.strftime('%d-%m-%Y')
```

Convert to numeric values

```python
lotto['Draw#'] = lotto['Draw#'].astype('int')
lotto['Power Ball'] = lotto['Power Ball'].astype('int')
lotto.Multiplier = lotto['Multiplier'].astype('int')
lotto['Jackpot'] = pd.to_numeric(lotto['Jackpot'].replace('[\$,]', '', regex=True))
lotto.Wins = lotto['Wins'].astype('int')
```
