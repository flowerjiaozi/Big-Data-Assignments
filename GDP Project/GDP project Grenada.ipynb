```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
grenada_df=pd.read_csv("Grenada GDP by sector.csv")
```

```python
grenada_df.head()
```

```python
grenada_df = grenada_df[['Sector','2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']]
```

```python
#Removing First Column
grenada_temp=grenada_df[1:]
grenada_temp.head()
```

```python
grenada_temp=grenada_temp.T
##removing the indexes
grenada_temp.reset_index(inplace=True)
print(grenada_temp.columns)
```

```python
grenada_temp
```

```python
years=grenada_temp['index']
years=pd.DataFrame(years)
years.columns=["Years"]
years=years[1:]
years.T
```

```python
grenada_temp=grenada_temp.drop(['index'],axis=1)
grenada_temp.T
grenada_temp.columns=grenada_temp.iloc[0]
```

```python
grenada_temp
grenada_temp["Year"]=years
grenada_temp.T
grenada_temp.head()
```

```python
grenada_clean=grenada_temp[1:]
grenada_clean = grenada_clean[['Year','Agriculture, Livestock and Forestry', 'Mining & Quarrying', 'Manufacturing', 'Electricity & Water', 'Wholesale & Retail Trade', 'Hotels & Restaurants', 'Transport, Storage and Communications', 'Financial Intermediation', 'Real Estate, Renting and Business Activities', 'Health and Social Work', 'Other Community, Social & Personal Services']]
grenada_clean.head()
```

```python
numeric_cols = grenada_clean.columns.drop('Year')
grenada_clean[numeric_cols] = grenada_clean[numeric_cols].astype(float)
```

```python
grenada_clean.head()
```

```python
# Set the figure size
plt.figure(figsize=(12, 10))

# Define a list of colors for the lines
colors = ['blue', 'green', 'red', 'purple', 'orange', 'yellow', 'cyan', 'magenta', 'lime', 'pink', 'brown']

# Plotting the graph with different colors for each line
for col, color in zip(grenada_clean.columns[1:], colors):
    plt.plot(grenada_clean['Year'], grenada_clean[col], marker='o', label=col, color=color)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Values in EC$ Million')
plt.title('Sector Contribution to Grenada GDP')

# Place the legend outside the plot area
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize='small')

plt.grid(True)  # Add grid lines
plt.xticks(grenada_clean['Year'])  # Set x-axis ticks to match years
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
# Plotting the scatter plot
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

# Plotting the Manufacturing sector on the first subplot
ax1.scatter(grenada_clean['Year'], grenada_clean['Transport, Storage and Communications'], marker='o', color='green', label='Transport, Storage and Communications', zorder=2)
ax1.set_title('Transport, Storage and Communications Sector')
ax1.set_xlabel('Year')
ax1.set_ylabel('Values in EC$ Million')
ax1.legend()
ax1.grid(True, zorder=1)

# Plotting the Trade and repairs sector on the second subplot
ax2.scatter(grenada_clean['Year'], grenada_clean['Real Estate, Renting and Business Activities'], marker='o', color='#FF1493' , label='Real Estate, Renting and Business Activities', zorder=2)
ax2.set_title('Real Estate, Renting and Business Activities Sector')
ax2.set_xlabel('Year')
ax2.set_ylabel('Values in EC$ Million')
ax2.legend()
ax2.grid(True, zorder=1)

# Plotting the Mining and quarrying sector on the third subplot
ax3.scatter(grenada_clean['Year'], grenada_clean['Wholesale & Retail Trade'], marker='o', color='orange', label='Wholesale & Retail Trade', zorder=2)
ax3.set_title('Wholesale & Retail Trade Sector')
ax3.set_xlabel('Year')
ax3.set_ylabel('Values in EC$ Millions in EC$ Million')
ax3.legend()
ax3.grid(True, zorder=1)


# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
